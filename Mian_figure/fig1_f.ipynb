```python
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
adata11=sc.read_h5ad('/data1/fengyu/methods/zy/P5_D06053D2_ananlysis/Latest_analysis_6.12/readid4/Final_D06053D2_after_processed_cellbin.h5ad')
```

```python
df=adata11.obs
tissue_outline=pd.DataFrame(adata11.uns['tissue_outline'],columns=['x','y'])
plt.style.use('dark_background')
# groupbynuniquetype
df['x'] = df['x'] //50*50
df['y'] = df['y'] //50*50
result_df = df.groupby(['x', 'y'])['IGHcdr3aa'].nunique().reset_index()
result_df.columns = ['x', 'y', 'counts']
plt.figure(figsize=(4.5, 4))
plt.grid(False)
sns.scatterplot(data=result_df[result_df['counts']!=0],x='x',y='y',hue='counts',s=5,edgecolor=None,marker = 's',
                palette='hot')
sns.scatterplot(data=tissue_outline,x='x',y='y',color='white',s=2,edgecolor=None)
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5),markerscale=3)
plt.savefig('./fig1e_right_IGH_overview.png',dpi=600, bbox_inches='tight')
#plt.show()
```
