# Sup6C

```python
batch_categories = ["D06047A2", "D06047C3", "D06047D4", "D06047E1", "D06047E2", "D06047F6", "D06050A2", "D06050C2", "D06050D4", "D06050E4", "D06053D2"]
```

```python
for i in range(11):
    ls=['#1f77b4', #Tcell
     '#ff7f0e', #bcell
     'lightgrey', #cancercell
     '#17becf', #ciliated cells
     '#9467bd', #endothelia
     '#8c564b', #epi
     '#e377c2', #fib
     '#2ca02c', #mast
     '#bcbd22', #myeloid
     '#d62728'] #plasma
    batch = batch_categories[i]
    data = adata[adata.obs['batch'] == batch]
    data.uns['anno_colors']=ls
    # 假设缺失值表现为 NaN（如果是空字符串先替换成 NaN）
    tmp = data[~data.obs['anno'].isna()].copy()

    sc.pl.umap(tmp,
           color=['anno'],
           s=2,
           palette=data.uns['anno_colors'],   # 你之前设好的颜色
           save=f'supfig6c{batch}.png')       # 注意字符串拼写
```
